```python
import gensim.downloader as api
model = api.load("glove-wiki-gigaword-50")
print(type(model))
```

```python
print(model["movie"])
print(model.similarity("movie","film"))
print(model.most_similar("fridge",topn=5))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 
words=["film","movie","car","bus","cat","dog"]
word_vector = [model[word] for word in words]
pca = PCA(n_components=2)
word_vector_2d = pca.fit_transform(word_vector)
plt.scatter(word_vector_2d[:,0],word_vector_2d[:,1])
for word,(x,y) in zip(words,word_vector_2d):
    plt.annotate(word,(x,y))
```
