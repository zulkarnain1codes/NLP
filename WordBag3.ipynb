```python
from nltk.tokenize import word_tokenize
import string
def preprocessor(text):
    text.lower()
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in string.punctuation]
    return " ".join(tokens)
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
reviews = [
    "The product is fantastic! It works like a charm.",
    "I hated the product. It broke after one use.",
    "Product was okay, not the best, but fine overall."
]
vectorizer = CountVectorizer()
```

```python
import numpy as np
import matplotlib.pyplot as plt
cleaned_review = [preprocessor(review) for review in reviews]
X=vectorizer.fit_transform(cleaned_review)
word_count = np.sum(X.toarray(),axis=0)
words = vectorizer.get_feature_names_out()
plt.bar(words,word_count)
plt.xticks(rotation=90)
```
